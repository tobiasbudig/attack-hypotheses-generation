```python
#imports
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#read partitioned Data for training (80% training data)
prob_table = pd.read_excel("../Data for algorithms.xlsx",sheet_name = 'X')

# read partitioned Data for testing (20% test data)
y_test = pd.read_excel("../Data for algorithms.xlsx", sheet_name="Y", index_col="ID")
y_test = y_test.drop(['created'], axis = 1).T

prob_table
```

```python
#prepare training data for simulation, output: techniques and their probabilities
prob_table = prob_table.drop(range(2, prob_table.shape[0]))
prob_table
```

```python

prob_table = prob_table.drop(columns = ["Unnamed: 0", "Unnamed: 1"])
prob_table = prob_table.transpose()

df_prob_table = pd.DataFrame(prob_table)
df_prob_table.columns=['ID', 'prop']

matrix_start = pd.read_excel("./attack-hypotheses-generation/Data for algorithms.xlsx", sheet_name = 'Basis').set_index('ID')
#df_prob_table["P(ID)"] = df_prob_table["prop"] * 100
#df_prob_table = df_prob_table.drop(["prop"], axis=1)

print(df_prob_table['prop'].sum())

df_prob_table
```

```python
# constants here, average techniques: 11, deviation of 6 techniques, 100 simulation
mu_number_technique = 12.0
sigma_number_technique = 9.0

number_of_simulations = 100
```

```python
#get random number of techniques from normal distribution
def generate_number_of_techniques(mu, sigma, dimension):
    n = np.random.normal(mu, sigma, dimension)
    n[ n <= 0] = 0
    return np.rint(n).astype(int)
```

```python
#generate defined amount of simulations, 12 techniques (average) per simulation

my_list_set = []
number_set= 12

for i in range(1,number_of_simulations+1):
    new_list_set= np.random.choice(df_prob_table["ID"], number_set, p=df_prob_table["prop"]).tolist()

    my_list_set.append(new_list_set)
    
my_list_set
```

```python
#generate defined amount of simulations, different amount of techniques per simulation 

my_list = []

for i in range(1,number_of_simulations+1):
    number= generate_number_of_techniques(mu_number_technique, sigma_number_technique, 1)
    
    new_list= np.random.choice(df_prob_table["ID"], number, p=df_prob_table["prop"]).tolist()

    my_list.append(new_list)
    
my_list
```

```python
#write simulation in known matrix form to compare
matrix = matrix_start.fillna(int(0))

for column in matrix: 
    software = my_list[column - 1]
    for technique in software: 
        matrix[column][technique] = 1
matrix
```

```python
#save simulation as a matrix
matrix.to_excel('result simulation as matrix.xlsx')
```

```python
# function to compare test data and simulation data with F1 score
    

def calculate_F_score(y_test: pd.DataFrame, predicted_sw: np.array, do_print=False):    
    #print(type(y_train), type(predicted_sw))
    assert y_test.shape[0] == predicted_sw.shape[0]
    
    ones = np.ones(len(y_test))

    occurences = np.array(predicted_sw).T @ ones
    
    F1_list = []
    Software_index = 0
    
    for software in y_test.columns:
        sum_s = y_test[software] @ predicted_sw

        precision = 0
        if occurences != 0:
            precision = sum_s/occurences
        
        number_of_techniques_y = y_test[software] @ ones
        recall = 0
        if number_of_techniques_y != 0:
            recall = sum_s/(number_of_techniques_y)
        F1 = 0
        if (precision + recall) != 0:
            F1 = 2 * precision* recall /(precision + recall)
            
        F1_list.append(F1)
    
    return max(F1_list), F1_list.index(max(F1_list)) , predicted_sw
```

```python
#Compare simulation and test data with F1 score
#Saving max_F1 and index in Dataframe 

max_F1_list = []

for generated in matrix: 
    maximum, index, predicted = calculate_F_score(y_test, np.array(matrix[generated]))
    
    max_F1_list.append((maximum, index))
    
scores = pd.DataFrame(max_F1_list)
scores.to_excel('F1 scores_simulation.xlsx')
#scores
```
