```python
# constants here
mu_number_technique = 11.0
sigma_number_technique = 6.0

number_of_simulations = 100
```

```python
#1. get random number of techniques from normal distribution
def generate_number_of_techniques(mu, sigma, dimension):
    n = np.random.normal(mu, sigma, dimension)
    n[ n <= 0] = 0
    #n=11
    return np.rint(n).astype(int)
```

```python
#100 draws from the probabilities, 11 techniques
#range(101)
my_list_set = []
number_set= 11

for i in range(1,number_of_simulations+1):
    new_list_set= np.random.choice(df_prob_table["ID"], number_set, p=df_prob_table["prop"]).tolist()

    my_list_set.append(new_list_set)
    
my_list_set
```

```python
#draws from the probabilities, different number of techniques
#range(101)
my_list = []

for i in range(1,number_of_simulations+1):
    number= generate_number_of_techniques(mu_number_technique, sigma_number_technique, 1)
    
    new_list= np.random.choice(df_prob_table["ID"], number, p=df_prob_table["prop"]).tolist()

    my_list.append(new_list)
    
my_list
```

```python
#Liste in bekanntes Matrix-Format bringen
matrix = matrix.fillna(int(0))

for column in matrix: 
    software = my_list[column - 1]
    for technique in software: 
        matrix[column][technique] = 1
matrix
```

```python
# read in the y_train data set and the predicted software

def calculate_F_score(y_test: pd.DataFrame, predicted_sw: np.array, do_print=False):    
    #print(type(y_train), type(predicted_sw))
    assert y_test.shape[0] == predicted_sw.shape[0]
    
    ones = np.ones(len(y_test))

    occurences = np.array(predicted_sw).T @ ones
    
    F1_list = []
    Software_index = 0
    
    for software in y_test.columns:
        sum_s = y_test[software] @ predicted_sw

        precision = 0
        if occurences != 0:
            precision = sum_s/occurences
        
        number_of_techniques_y = y_test[software] @ ones
        recall = 0
        if number_of_techniques_y != 0:
            recall = sum_s/(number_of_techniques_y)
        F1 = 0
        if (precision + recall) != 0:
            F1 = 2 * precision* recall /(precision + recall)
            
        F1_list.append(F1)
    
    return max(F1_list), F1_list.index(max(F1_list)) , predicted_sw
```

```python
# Saving max F1 and index in Dataframe 

max_F1_list = []

for generated in matrix: 
    maximum, index, predicted = calculate_F_score(y_test, np.array(matrix[generated]))
    
    max_F1_list.append((maximum, index))
    
pd.DataFrame(max_F1_list)
```
