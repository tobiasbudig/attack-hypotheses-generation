```python
#Do the necessary imports
import numpy as np
import os
from kneed import KneeLocator
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans, k_means
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
 
#Import Data, droping Summe and ID
matrix = pd.read_excel("./Data with time stemps reduced.xlsx", "Matrix without time").drop(columns=['Summe', 'ID'])
```

```python
x_train = matrix#.values

no_of_clusters = range(2, 50)
inertia = []

for f in no_of_clusters:
    kmeans = KMeans(n_clusters = f, random_state = 2).fit(x_train)
    u = kmeans.inertia_
    inertia.append(u)
    label = kmeans.fit_predict(x_train)
    #print(label)
    #print("The inertia for: ", f, "Clusters is: ", u)
```

```python
kmeans = KMeans(n_clusters = 23, random_state = 2).fit(x_train)
u = kmeans.inertia_
inertia.append(u)
label = kmeans.fit_predict(x_train)
print(label)
#print("The inertia for: ", f, "Clusters is: ", u)
```

```python
#plotting elbow curve
fig, (ax1) = plt.subplots(1, figsize = (16,6))
xx = np.arange(len(no_of_clusters))
ax1.plot(xx, inertia)
ax1.set_xticks(xx)
ax1.set_xticklabels(no_of_clusters, rotation = 'vertical')
plt.xlabel('Number of clusters')
plt.ylabel('Intertia Score')
plt.title("Intertia Plot per k")

#caldulating elbow
kl = KneeLocator(range(2, 50), inertia, curve="convex", direction="decreasing")
kl.elbow
```
